# Simple linear regression
This is a simple linear regression for predicting how age affects insurance charges

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

### Import the data

```python
data = pd.read_csv('insurance.csv')
```

### Check the data headings

```python
data.head()
```

### Scatter plot of the age by charges
The iloc function allows us to select rows from the dataframe for plotting 

```python
# select all the rows of the 1st column 
x = data.iloc[:,:1].values 

# select all the rows of the 7th column 
y = data.iloc[:,6].values

# Plot the scatter plot
plt.scatter(x,y, marker = '*', color = '#483D8B')

# Label both axis of the scatter plot
plt.xlabel('Age')
plt.ylabel('Charges')

# Add a title
plt.title('Scatter plot of Age by Charges')

plt.show()
```

### Fit the model

```python
model = LinearRegression()
model.fit(x,y)
```

### Scatter plot with predictions

```python
y_pred = model.predict(x)

# Plot the scatter plot
plt.scatter(x,y, marker = '*', color = '#483D8B')
plt.plot(x,y_pred, color = 'r')

# Label both axis of the scatter plot
plt.xlabel('Age')
plt.ylabel('Charges')

# Add a title
plt.title('Scatter plot showing Line of Best Fit')

# Enable grid lines
plt.grid(linestyle = '--')

plt.show()
```

### Make predictions
Prediction of the insurance charges for age 43

```python
# Age to be predicted
unk_x = [[43]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = model.predict(x_pred)

# Plot the scatter plot
plt.scatter(x,y, marker = '*', color = '#483D8B')
plt.plot(x_pred, y_pred, color = 'r')

# Label both axis of the scatter plot
plt.xlabel('Age')
plt.ylabel('Charges')

# Add a title
plt.title('Scatter plot with Best-Fit Line')

# Enable grid lines
plt.grid(linestyle = '--')

plt.show()

print("Charges should be:", model.predict(unk_x))
```
